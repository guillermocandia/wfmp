```python
import http.client
import json
import pandas as pd

from pprint import pprint
```

```python
conn = http.client.HTTPSConnection("api.warframe.market")

headers = {
    'platform': "pc",
    'crossplay': "true"
}
```

```python
conn.request("GET", "/v2/items", headers=headers)

response = conn.getresponse()

if response.status == 200: 
    data = response.read().decode("utf-8")
    json_data = json.loads(data)
    print(len(json_data["data"]))
else:
    print(f"Request failed with status {response.status}: {response.reason}")
conn.close()
```

```python
items = json_data["data"]
pprint(len(items))
```

```python
filtered_items = list(filter(lambda x: "arcane_enhancement" in x["tags"] or "mod" in x["tags"], items))
pprint(len(filtered_items))
pprint(filtered_items[:2])
```

```python
def parse_rarity(tags):
    if "common" in tags:
        return "common"
    elif "uncommon" in tags:
        return "uncommon"
    elif "rare" in tags:
        return "rare"
    elif "legendary" in tags:
        return "legendary"
    else:
        return None
```

```python
def parse_item(item):
    return {
        "id": item["id"],
        "slug": item["slug"],
        "name": item["i18n"]["en"]["name"],
        "icon": item["i18n"]["en"]["icon"],
        "max_rank": item.get("maxRank"),
        "type": "arcane" if "arcane_enhancement" in item["tags"] else "mod",
        "rarity": parse_rarity(item["tags"])
    }
```

```python
items = [parse_item(x) for x in filtered_items]
pprint(len(items))
pprint(items[:2])
```

```python
items = filter(lambda x: x["max_rank"]==None, items)
pprint(list(items))
```
